```python
def filter_by_market_logic(df):
    # 1. DHA Logic: 8 million to 10 million per marla
    dha_mask = (df['Location'] == 'DHA Defence')
    df = df[~(dha_mask & ((df['price_per_marla'] < 8000000) | (df['price_per_marla'] > 10000000)))]
    
    # 2. Johar/Bahria Logic: Slightly less than DHA (e.g., 5 million to 8 million)
    # Adjusting based on your 'slightly less' description
    mid_mask = (df['Location'].isin(['Johar Town', 'Bahria Town']))
    df = df[~(mid_mask & ((df['price_per_marla'] < 5000000) | (df['price_per_marla'] > 80000000)))]
    
    # 3. Others: 50k to 3 Lac per marla
    # We apply this to everything else
    other_mask = (~df['Location'].isin(['DHA Defence', 'Johar Town', 'Bahria Town']))
    df = df[~(other_mask & ((df['price_per_marla'] < 50000) | (df['price_per_marla'] > 3000000)))]
    
    return df

# Apply your specific market knowledge
df4 = filter_by_market_logic(df3)

print(f"Rows remaining after applying your market logic: {len(df4)}")
```

```python
def find_price_anomalies(df):
    anomalies = []
    # Group by Location and Area (Size)
    for (location, area), group in df.groupby(['Location', 'Area']):
        # Get all 2-bedroom houses in this group
        bed2 = group[group['Bedroom(s)'] == 2]
        # Get all 3-bedroom houses in this group
        bed3 = group[group['Bedroom(s)'] == 3]
        
        if not bed2.empty and not bed3.empty:
            max_price_2bed = bed2['Price'].max()
            min_price_3bed = bed3['Price'].min()
            
            # If the most expensive 2-bed is more than the cheapest 3-bed
            if max_price_2bed > min_price_3bed:
                anomalies.append(group)
                
    return pd.concat(anomalies) if anomalies else pd.DataFrame()

# Run the check on your cleaned data
price_anomalies = find_price_anomalies(df4)
price_anomalies.head(10)
```

```python
df5=df4.copy()
```

```python
# 1. Create a copy to keep your data safe
df_final = df5.copy()

# 2. Convert 'Type' to binary (1 for House, 0 for Flat)
# We use 'str.contains' just in case there are spaces
df_final['Type'] = df_final['Type'].apply(lambda x: 1 if str(x).strip() == 'House' else 0)

# 3. DROP the text columns that are causing the error
# We remove 'Location' because we already have the dummy columns for it
columns_to_drop = ['Location']
if 'price_per_marla' in df_final.columns:
    columns_to_drop.append('price_per_marla')

df_final = df_final.drop(columns_to_drop, axis='columns', errors='ignore')

# 4. Now convert to int (True becomes 1, False becomes 0)
df_final = df_final.astype(int)

print("Success! Dataframe is now all numeric.")
print(f"Final Shape: {df_final.shape}")
df_final.head()
```

```python
# 1. Get the dummy variables for the locations in your filtered data
dummies = pd.get_dummies(df4.Location)

# 2. Concatenate the dummies with your cleaned numeric columns
# We drop 'other' to avoid the dummy variable trap
df_ready = pd.concat([df_final, dummies.drop('other', axis='columns', errors='ignore')], axis='columns')

# 3. Final check: You should see 'Type', 'Area', etc. followed by 100+ location columns
print(f"Final shape for training: {df_ready.shape}")
df_ready.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# X is our features (everything except Price)
X = df_ready.drop('Price', axis='columns')

# y is our target (Price)
y = df_ready['Price']

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)

# Initialize and Train the model
lr_clf = LinearRegression()
lr_clf.fit(X_train, y_train)

# Check the accuracy
score = lr_clf.score(X_test, y_test)
print(f"Model Accuracy (R-squared): {score * 100:.2f}%")
```

```python
import numpy as np

def predict_price(location, area, bath, bed, is_house):    
    # 1. Find the column index for the location
    try:
        loc_index = np.where(X.columns == location)[0][0]
    except:
        loc_index = -1

    # 2. Create an array of 0s
    x_input = np.zeros(len(X.columns))
    x_input[0] = 1 if is_house == 'House' else 0
    x_input[1] = area
    x_input[2] = bath
    x_input[3] = bed
    
    if loc_index >= 0:
        x_input[loc_index] = 1

    # 3. FIX: Convert the array into a DataFrame with feature names
    x_df = pd.DataFrame([x_input], columns=X.columns)

    # 4. Predict using the DataFrame instead of the array
    prediction = lr_clf.predict(x_df)[0]
    return f"Estimated Price: {prediction:,.0f} PKR"
```
